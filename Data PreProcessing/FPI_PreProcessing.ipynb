```python
# prompt: I need to import a bunch of csv filed from my computer. There are fpi statistics for college football from the year 2005-2024. Each year is its own file (example "2024_fpi.csv"). Please write code to import all of these files

from google.colab import files, drive
import pandas as pd
import io
import os
```

```python
# Mount Google Drive
drive.mount('/content/drive')

# Define the folder path within Google Drive
folder_path = '/content/drive/My Drive/CFB_Model/Raw Data/fpi/'

# List all files in the folder
files = os.listdir(folder_path)

# Load all CSV files within the folder
dataframes = {}

for file_name in files:
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        dataframes[file_name] = pd.read_csv(file_path)

# Show the keys of the loaded DataFrames
print("Loaded dataframes:", dataframes.keys())
```

```python
# Assume all dataframes should have the same columns
if dataframes:
  first_df_columns = dataframes[list(dataframes.keys())[0]].columns
  all_match = True
  for df_name, df in dataframes.items():
    if not df.columns.equals(first_df_columns):
      all_match = False
      print(f"Dataframe '{df_name}' has different columns.")
      break

  if all_match:
    print("All dataframes have matching columns.")
else:
  print("No dataframes were uploaded.")
```

```python
# prompt: concatonate the dataframes into a single dataframe and drop the "ResumeRanks RemainingStrengthOfSchedule" column from the concatonated dataframe

# Concatenate all dataframes into a single dataframe
concatenated_df = pd.concat(dataframes.values(), ignore_index=True)

# Drop the "ResumeRanks RemainingStrengthOfSchedule" column
concatenated_df = concatenated_df.drop(columns=["ResumeRanks RemainingStrengthOfSchedule","ResumeRanks StrengthOfRecord","ResumeRanks AverageWinProbability","ResumeRanks StrengthOfSchedule"], errors='ignore')

print(concatenated_df)
```
